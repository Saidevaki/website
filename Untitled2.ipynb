```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import euclidean
from sklearn.preprocessing import StandardScaler
import random

class CanopyClustering:
    def __init__(self, t1, t2):
        """ Initialize Canopy Clustering with loose (t1) and tight (t2) thresholds """
        self.t1 = t1
        self.t2 = t2
        self.canopies = []

    def fit(self, data):
        """ Perform Canopy Clustering on Features (Columns) Instead of Rows """
        num_features = data.shape[1]  # Correctly get the number of features (87)
        remaining_features = set(range(num_features))  # Track feature indices

        print(f"Dataset Shape: {data.shape}")  # Debugging
        print(f"Total Features: {num_features}")  # Debugging

        while remaining_features:
            center = random.choice(list(remaining_features))  # Pick a random feature
            remaining_features.remove(center)  # Remove selected feature

            canopy = {center}  # Initialize a new canopy
            to_remove = set()

            for feature in remaining_features:
                dist = euclidean(data[:, center], data[:, feature])  # Compute feature-wise distance
                if dist < self.t1:
                    canopy.add(feature)
                if dist < self.t2:
                    to_remove.add(feature)

            self.canopies.append(canopy)
            remaining_features -= to_remove  # Remove tightly clustered features

    def get_canopies(self):
        """ Return the formed canopies """
        return self.canopies

# Load dataset
data = pd.read_csv("/content/dataset_phishing.csv")

# Drop non-feature columns
X = data.drop(columns=['url', 'status'])

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Do not transpose here

# Run Canopy Clustering
t1, t2 = 1.5, 0.8  # Adjust thresholds as needed
canopy = CanopyClustering(t1, t2)
canopy.fit(X_scaled)

# Fetch canopies
canopies = canopy.get_canopies()

# Feature selection: Choose one representative feature per canopy
selected_features = []
for canopy_group in canopies:
    representative = list(canopy_group)[0]  # Pick first feature from each canopy
    selected_features.append(X.columns[representative])

# Print selected features
print("Selected Features:", selected_features)
print("No.of selected Features : ",len(selected_features))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("/content/dataset_phishing.csv")  # Change to your actual dataset file

# Check available columns
print("Available Columns:", df.columns)

# Define target column
target_column = "status"

if target_column not in df.columns:
    raise ValueError(f"Column '{target_column}' not found. Available columns: {df.columns}")

# Convert categorical target variable to numeric
label_encoder = LabelEncoder()
df[target_column] = label_encoder.fit_transform(df[target_column])

# **Convert Categorical Features to Numeric**
for col in df.columns:
    if df[col].dtype == 'object':  # If column is categorical (string)
        df[col] = label_encoder.fit_transform(df[col])  # Convert to numbers

# Separate features and target
X = df.drop(columns=[target_column])
y = df[target_column]

# **Canopy Clustering for Feature Selection (Using KMeans as Approximation)**
n_clusters = 10  # Adjust based on dataset size
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
kmeans.fit(X)

# Select the most important features based on cluster centroids
feature_importance = np.argsort(np.linalg.norm(kmeans.cluster_centers_, axis=0))[-15:]  # Select top 15 features
X_selected = X.iloc[:, feature_importance]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Train Gradient Boosting Classifier
gbc = GradientBoostingClassifier(n_estimators=500, learning_rate=0.05, max_depth=5, random_state=42)
gbc.fit(X_train, y_train)

# Predictions
y_train_pred = gbc.predict(X_train)
y_test_pred = gbc.predict(X_test)

# Evaluate model
acc_train = metrics.accuracy_score(y_train, y_train_pred)
acc_test = metrics.accuracy_score(y_test, y_test_pred)
f1_train = metrics.f1_score(y_train, y_train_pred, average='macro')
f1_test = metrics.f1_score(y_test, y_test_pred, average='macro')
recall_train = metrics.recall_score(y_train, y_train_pred, average='macro')
recall_test = metrics.recall_score(y_test, y_test_pred, average='macro')
precision_train = metrics.precision_score(y_train, y_train_pred, average='macro')
precision_test = metrics.precision_score(y_test, y_test_pred, average='macro')

# Print Metrics
print(f"Gradient Boosting Classifier - Accuracy: Train={acc_train:.3f}, Test={acc_test:.3f}")
print(f"Gradient Boosting Classifier - F1 Score: Train={f1_train:.3f}, Test={f1_test:.3f}")
print(f"Gradient Boosting Classifier - Recall: Train={recall_train:.3f}, Test={recall_test:.3f}")
print(f"Gradient Boosting Classifier - Precision: Train={precision_train:.3f}, Test={precision_test:.3f}")
```
